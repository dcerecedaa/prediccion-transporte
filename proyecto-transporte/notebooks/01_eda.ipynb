```python
import pandas as pd
from pathlib import Path


def resolver_data_dir() -> Path:
    # Path.cwd() devuelve el directorio actual (notebooks/).
    # .parent sube un nivel hasta la raíz del proyecto.
    # El operador / construye la ruta de forma compatible.
    return Path.cwd().parent / "data" / "raw"


data_dir = resolver_data_dir()

# Bucle para detectar registros con menos de 50 viajeros diarios por línea
for archivo in sorted(data_dir.glob("demandadialinea_2[0-9]*.csv")):
    df = pd.read_csv(archivo, sep=";")

    # Filtramos registros con menos de 50 viajeros diarios por línea
    sospechosos = df[df["TotalViajeros"] < 50]

    if len(sospechosos) > 0:
        print(f"\n{'='*50}")
        print(f"ARCHIVO: {archivo.name}")
        print(f"Registros con menos de 50 viajeros: {len(sospechosos)}")

        # Agrupamos por línea para ver si los valores bajos son sistemáticos
        # (líneas de baja demanda real) o aleatorios (posibles errores de registro)
        print(sospechosos.groupby("Linea")["TotalViajeros"]
              .agg(["count", "min", "max", "mean"])
              .sort_values("min")
              .head(20))
```
